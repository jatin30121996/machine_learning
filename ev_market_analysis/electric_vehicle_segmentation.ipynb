```python
data1 = dffs[["rating", "Visual Appeal", "Reliability", "Performance", "Service Experience", "Extra Features", "Comfort", "Maintenance cost", "Value for Money"]]
```

```python
data1
```

```python
data1.isna().sum()
```

```python
data1.fillna(0., inplace=True)
```

```python
data1.describe()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
data1.isna().sum()
```

```python
scaler = StandardScaler()
scaled_data1 = scaler.fit_transform(data1)
```

```python
pca = PCA(random_state=123)
pca.fit(scaled_data1)
```

```python
fitted_data = pca.transform(scaled_data1)
```

```python
fitted_df = pd.DataFrame(fitted_data, columns=["PC_"+str(x+1) for x in range(len(data1.columns))])
fitted_df
```

```python
pca_summary = pd.DataFrame({
    "Standard Deviation": fitted_data.std(),
    "Proportion of Variance": pca.explained_variance_ratio_,
    "cumulative_proportion": pca.explained_variance_ratio_.cumsum()
}, index=fitted_df.columns)
```

```python
pca_summary
```

```python
fl = pd.DataFrame(pca.components_, columns=fitted_df.columns, index=data1.columns)
```

```python
fl.style.background_gradient(cmap="Reds")
```

```python
plt.figure(figsize=(12, 10))
index_names = fl.index
plt.scatter(fitted_data[:, 0], fitted_data[:, 1], color="grey", alpha=0.5)
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)
for i in range(loadings.shape[0]):
    plt.arrow(0, 0, loadings[i, 0], loadings[i, 1], color='red', alpha=0.7, head_width=0.05, head_length=0.1)
    plt.text(loadings[i, 0], loadings[i, 1], index_names[i], fontsize=12, ha='center', color = 'red', va = 'center_baseline')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Plot with Projection Axes')
plt.grid(True)
plt.xlim(-3, 3) 
plt.ylim(-3, 3) 
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.show()
```

```python
data_clusters = [KMeans(n_clusters=i, random_state=123).fit(fitted_df) for i in range(2, 10)]
```

```python
sns.lineplot(x=[i for i in range(2, 10)], y = [data_cluster.inertia_ for data_cluster in data_clusters])
plt.title("Best fit with number of titles")
```

```python
model = KMeans(n_clusters=6, random_state=123).fit(fitted_df)
```

```python
data_clust = pd.DataFrame(scaled_data1, columns=data1.columns)
data_clust["cluster"] = model.labels_
```
