```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
MD_x_scaled = scaler.fit_transform(md_x)
MD_x_scaled
```

```python
pca = PCA()
MD_pca = pca.fit_transform(MD_x_scaled)
MD_pca
```

```python
MD_pca_df = pd.DataFrame(MD_pca, columns=[f'PC{i+1}' for i in range(MD_pca.shape[1])])
explained_variance = pca.explained_variance_
explained_variance_ratio = pca.explained_variance_ratio_
```

```python
summary_df = pd.DataFrame({
    'Principal Component': [f'PC{i+1}' for i in range(len(explained_variance))],
    'Explained Variance': explained_variance,
    'Explained Variance Ratio': explained_variance_ratio
})
summary_df['Explained Variance'] = summary_df['Explained Variance'].round(2)
summary_df['Explained Variance Ratio'] = summary_df['Explained Variance Ratio'].round(2)
print(summary_df)
```

```python
MD_pca_df = pd.DataFrame(MD_pca, columns=[f'PC{i+1}' for i in range(MD_pca.shape[1])])
print(MD_pca_df.round(1))
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(MD_pca[:, 0], MD_pca[:, 1], color='grey', alpha=0.5)
loadings = pca.components_.T * np.sqrt(pca.explained_variance_)
for i in range(loadings.shape[0]):
    plt.arrow(0, 0, loadings[i, 0], loadings[i, 1], 
              color='red', alpha=0.7, 
              head_width=0.05, head_length=0.1)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Plot with Projection Axes')
plt.grid(True)
plt.xlim(-3, 3) 
plt.ylim(-3, 3) 
plt.axhline(0, color='black',linewidth=0.5, ls='--')
plt.axvline(0, color='black',linewidth=0.5, ls='--')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
np.random.seed(1234)
```

```python
best_k = None
best_model = None
best_score = -1
```

```python
models = {}
silhouette_scores = []
```

```python
for k in range(2, 9):
    model = KMeans(n_clusters=k, n_init=10, random_state=1234)
    model.fit(md_x)
    score = silhouette_score(md_x, model.labels_)
    silhouette_scores.append(score)
    if score > best_score:
        best_score = score
        best_k = k
        best_model = model
    
    models[k] = model
```

```python
def relabel_clusters(labels):
    unique_labels = np.unique(labels)
    label_mapping = {label: idx for idx, label in enumerate(sorted(unique_labels))}
    relabeled = np.vectorize(label_mapping.get)(labels)
    return relabeled
```

```python
if best_model:
    relabeled_labels = relabel_clusters(best_model.labels_)
```

```python
results = pd.DataFrame(md_x)
results['Cluster'] = relabeled_labels
```

```python
print(f"Best number of clusters: {best_k}")
print(results.head())
```

```python
plt.figure(figsize=(8, 5))
plt.plot(range(2, 9), silhouette_scores, marker='o')
plt.title('Silhouette Scores for Different Numbers of Clusters')
plt.xlabel('Number of Segments (Clusters)')
plt.ylabel('Silhouette Score')
plt.xticks(range(2, 9))
plt.grid()
plt.show()
```

```python
from sklearn.metrics import adjusted_rand_score
from sklearn.utils import resample
```

```python
cluster_range = range(2, 9)
ari_results = pd.DataFrame(index=cluster_range, columns=range(100))
```

```python
nboot = 100
x = pd.DataFrame(index=range(1453))
for k in cluster_range:
    for boot in range(nboot):
        md_x_boot = resample(md_x, random_state=boot)
        
        model = KMeans(n_clusters=k, n_init=10, random_state=boot)
        x["cluster"] = model.fit_predict(md_x_boot)
        
        ari_results.loc[k, boot] = model.labels_
```

```python
ari_means = []
for k in cluster_range:
    original_labels = ari_results.loc[k, 0]
    ari_scores = [adjusted_rand_score(original_labels, ari_results.loc[k, boot]) for boot in range(nboot)]
    ari_means.append(np.mean(ari_scores))
```

```python
plt.figure(figsize=(8, 5))
plt.plot(cluster_range, ari_means, marker='o')
plt.title('Adjusted Rand Index for Different Numbers of Clusters')
plt.xlabel('Number of Segments (Clusters)')
plt.ylabel('Adjusted Rand Index')
plt.xticks(cluster_range)
plt.grid()
plt.show()
```

```python
cluster_label = 4
cluster_data = x[x['cluster'] == cluster_label]
plt.figure(figsize=(8, 5))
plt.hist(x["cluster"], range=(0, 1), alpha=0.7)
plt.title(f'Histogram of Data Points in Cluster {cluster_label}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.xlim(0, 1)
plt.grid()
plt.show()
```

```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import log_loss

np.random.seed(1234)

results = {}
for k in range(2, 9):
    model = GaussianMixture(n_components=k, random_state=1234, n_init=10)
    model.fit(md_x)
    results[k] = model


MD_m28 = pd.DataFrame({
    'k': list(results.keys()),
    'AIC': [model.aic(md_x) for model in results.values()],
    'BIC': [model.bic(md_x) for model in results.values()],
})

print(MD_m28)

```

```python
plt.figure(figsize=(10, 6))
plt.plot(MD_m28['k'], MD_m28['AIC'], marker='o', label='AIC', color='blue')
plt.plot(MD_m28['k'], MD_m28['BIC'], marker='o', label='BIC', color='orange')
plt.title('Information Criteria for Mixture Models')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Value of Information Criteria')
plt.xticks(MD_m28['k'])
plt.grid()
plt.legend()
plt.show()
```

```python
coefficients = model.coef_
intercept = model.intercept_

# Create a summary-like output
summary = pd.DataFrame({'Feature': X.columns, 'Coefficient': coefficients})
summary['Significance'] = np.where(summary['Coefficient'] != 0, 'Yes', 'No')

# Display the summary
print("Summary of Regression Model:")
print(summary)
```
